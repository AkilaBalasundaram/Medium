```python
#Building model using statsmodel, for the detailed statistics
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train_RFE[col]
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
lm = sm.OLS(y_train,X_train_rfe).fit() 
print(lm.summary())
```

#### Checking VIF

checking VIF parameter is extremely important since it gives us how correlated are the variables between themselves.This ensures the stability of model.

**Adjusted Rsquare is 84% which is a good score ,
Looking at the P- values most of them becomes insignificant, atemp which could explain about 40% of variability has now become insignificant
which indicate there are multicollinearity and the model is unstable, let  us supplement this with VIF information and decide
which variables to keep and which variables to drop.**


```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
VIF_train=X_train[col]
vif = pd.DataFrame()
vif['Features'] = VIF_train.columns
vif['VIF'] = [variance_inflation_factor(VIF_train.values, i) for i in range(VIF_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
```

### Fine tuning the model

let us consider Pvalue and VIF of the features to fine tune model, follow below steps for fine tuning.

- **High P value,High VIF - drop first**
- **Low Pvalue , low VIF - keep the feature**
- **Low Pvalue , High VIF - drop and check the performance to take the decision**
- **High Pvalue , Low VIF - drop and check the performance to take the decision**
