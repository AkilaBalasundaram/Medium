#  EDA ON LOAN APPLICATION RISKS

## Introduction
This case study aims to give you an idea of applying EDA in a real business scenario. In this case study, apart from applying the techniques that you have learnt in the EDA module, you will also develop a basic understanding of risk analytics in banking and financial services and understand how data is used to minimise the risk of losing money while lending to customers.

## Business Understanding
The loan providing companies find it hard to give loans to the people due to their insufficient or non-existent credit history. Because of that, some consumers use it as their advantage by becoming a defaulter. Suppose you work for a consumer finance company which specialises in lending various types of loans to urban customers. You have to use EDA to analyse the patterns present in the data. This will ensure that the applicants capable of repaying the loan are not rejected.

 

When the company receives a loan application, the company has to decide for loan approval based on the applicant’s profile. Two types of risks are associated with the bank’s decision:

If the applicant is likely to repay the loan, then not approving the loan results in a loss of business to the company

If the applicant is not likely to repay the loan, i.e. he/she is likely to default, then approving the loan may lead to a financial loss for the company.

 

The data given below contains the information about the loan application at the time of applying for the loan. It contains two types of scenarios:

The client with payment difficulties: he/she had late payment more than X days on at least one of the first Y instalments of the loan in our sample,

All other cases: All other cases when the payment is paid on time.

 

When a client applies for a loan, there are four types of decisions that could be taken by the client/company):

Approved: The Company has approved loan Application

Cancelled: The client cancelled the application sometime during approval. Either the client changed her/his mind about the loan or in some cases due to a higher risk of the client he received worse pricing which he did not want.

Refused: The company had rejected the loan (because the client does not meet their requirements etc.).

Unused offer:  Loan has been cancelled by the client but on different stages of the process.

## Business Objectives
This case study aims to identify patterns which indicate if a client has difficulty paying their installments which may be used for taking actions such as denying the loan, reducing the amount of loan, lending (to risky applicants) at a higher interest rate, etc. This will ensure that the consumers capable of repaying the loan are not rejected. Identification of such applicants using EDA is the aim of this case study.

 

In other words, the company wants to understand the driving factors (or driver variables) behind loan default, i.e. the variables which are strong indicators of default.  The company can utilise this knowledge for its portfolio and risk assessment.

To develop your understanding of the domain, you are advised to independently research a little about risk analytics - understanding the types of variables and their significance should be enough).

### UNIVARIATE ANALYSIS

```python
def univar_ana2(variable,title_name,colorList):
    plt.figure(figsize=(10,25))
    plt.subplot(3,1,2)
    ax=(application_data[variable].value_counts(normalize=True)*100).plot.bar(color=colorList)
    plt.xticks(rotation=0)
    plt.title(title_name, fontdict = {"fontsize":20})
    for p in ax.patches:
            ax.annotate('{:.1f} %'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.50))
    plt.figure(figsize=(10,25))
    plt.subplot(3,1,2)
    ax=(defaulter[variable].value_counts(normalize=True)*100).plot.bar(color=colorList)
    plt.xticks(rotation=0)
    plt.title("DEFAULTERS : " + title_name, fontdict = {"fontsize":20})
    for p in ax.patches:
            ax.annotate('{:.1f} %'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.50))
    plt.show()        
    plt.figure(figsize=(10,25))
    plt.subplot(3,1,2)
    ax=(non_defaulters[variable].value_counts(normalize=True)*100).plot.bar(color=colorList)
    plt.xticks(rotation=0)
    plt.title("NON - DEFAULTER: "+ title_name, fontdict = {"fontsize":20})
    for p in ax.patches:
            ax.annotate('{:.1f} %'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.50))
    plt.show()
```

## Segmented analysis of EMPLOYMENT

```python
def segment_analysis(variable,title_name,colorlist):
    plt.figure(figsize=(10,25))
    plt.subplot(3,1,2)
    ax=(defaulter[variable].value_counts(normalize=True)*100).plot.bar(color=colorlist)
    plt.xticks(rotation=0)
    plt.title("DEFAULTERS : "+ title_name, fontdict = {"fontsize":20})
    for p in ax.patches:
            ax.annotate('{:.1f} %'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.50))
    plt.show()        
    plt.figure(figsize=(10,25))
    plt.subplot(3,1,2)
    ax=(non_defaulters[variable].value_counts(normalize=True)*100).plot.bar(color=colorlist)
    plt.xticks(rotation=0)
    plt.title("NON - DEFAULTER: " + title_name, fontdict = {"fontsize":20})
    for p in ax.patches:
            ax.annotate('{:.1f} %'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+0.50))
    plt.show()
```

```python
def segment_analysis2(variable):
    fig,(ax2,ax1) = plt.subplots(1,2,figsize=(4,4))
    sns.countplot(data=defaulter,x=variable,order = defaulter[variable].value_counts().sort_index().index,ax=ax1 )
    ax1.set_ylabel('Cases for defaulters')
    ax1.set_xlabel(variable)
    ax1.set_title('Distribution of Defaulters')
    ax1.set_xticklabels(ax1.get_xticklabels(),rotation=90,fontsize = 10, va='top', ha='right')
    
    # Adding percentage
    for p in ax1.patches:
        ax1.annotate('{:.0f}%'.format((p.get_height()/len(defaulter))*100), (p.get_x()+0.3, p.get_height()+100))
           
    sns.countplot(data=non_defaulters,x=variable,order = non_defaulters[variable].value_counts().sort_index().index ,ax=ax2)
    ax2.set_ylabel('Cases for NON-defaulters')
    ax2.set_xlabel(variable)
    plt.subplots_adjust(left=0.1,bottom=0.1, right=2, top=0.9, wspace=.4, hspace=.4)

    ax2.set_title('Distribution of NON-Defaulters')
    ax2.set_xticklabels(ax2.get_xticklabels(),rotation=90,fontsize = 10, va='top', ha='right')
    
    # Adding percentage 
    for p in ax2.patches:
        ax2.annotate('{:.0f}%'.format((p.get_height()/len(non_defaulters))*100), (p.get_x()+0.3, p.get_height()+100))
    
    plt.show()
```

```python
outlier_dict={}
for column in previous_data.describe().columns:
    diff=np.max(previous_data[column])-np.percentile(previous_data[column],75)
    maximum=np.max(previous_data[column])
    value_75=np.percentile(previous_data[column],75)
    outlier_dict[column]=[maximum,value_75,diff]
is_outlier_df=pd.DataFrame.from_dict(outlier_dict,orient='index',columns=['Maximum','75th_percentile','Difference'])
is_outlier_df
```

#### Recommendations

* *INCOME TYPE:* State servents/pensioners are less likely to default while unemployed/maternity faction of people are more likely to default
* *EDUCATION TYPE:*Clients with academic degree are repayers while clients with secondary education are more likely defaulers.    
* *GENDER :* Female clients are less likely to default than male client , according to this dataset(pleasekeep this is imbalanced dataset with respect to gender)
* *AGE GROUP:* clients with age group of 20-30 are more not the safest to approve loan , while  60+ are safest to sanction loans
* *Employment years:* clients whose employment experience is between 0-5 has paymment difficulties.
* *Marital status :* clients who are single/had civil marriage are more likely to default.while widow tend to repay on time.
* *Customer type :* Repeat customers are more reliable in terms of repayment compared to new customers.
* *Repeaters:* Customers whose previous application has been cancelled/refused has repayed more than 90% of time,so the cause of refusal/cancellation needs to be closely evaluated
* *Phone number :* Client who has provided work phone,and whose mobile was reachable has been more reliable.
* *Refusal/cancellations:* The client whose previous loan application was refused or cancelled , has repayed almost 90% of time
   the reson for refusal/cancellation has to be revisited or add extra check points, in order to not lose these good clients.
* *Number of children :* As number of children increases the client is more likely to default.
* *Number of Family members :* Same case as number of children ,with family members
